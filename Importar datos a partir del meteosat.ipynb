```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily
from geopy.geocoders import Nominatim
```

```python
#Obtenemos coordenadas
loc= Nominatim (user_agent="GetLoc")
getLoc= loc.geocode ("Ourense")
print(getLoc.address)
```

```python
#Obtenemos latitud y longitud
lat= getLoc.latitude
print("Latitude=",lat)
```

```python
long=getLoc.longitude
print("Longitude=",long)
```

```python
#Definir periodo de tiempo
start = datetime(1980,1,1)
end=datetime (2023,1,1)
```

```python
#Establecemos punto
Ourense= Point (lat,long)
```

```python
#Obtenemos datos diarios
datos1= Daily(Ourense,start,end)
datos1=datos1.fetch()
datos1
```

```python

datos1.head()
```

```python
datos1.index
```

```python
datos1['Año'] = datos1.index.year       #utilizamos la funcion .index (para ver la fecha( en mes,año y dia separado))                         
datos1['Mes'] = datos1.index.month
datos1['Dia'] = datos1.index.day 
datos1.sample(45, random_state=0)
```

```python
datos1=datos1[["Año","prcp"]]
datos1
```

```python
datos2=datos1.dropna() #eliminar datos nulos
datos2
```

```python
#Histograma precipitacion ourense desde 1980
```

```python
plt.bar(datos2['Año'], datos2['prcp'])
plt.xlabel('Año')
plt.ylabel('Precipitacion (mm)')
plt.title('Histograma de Precipitacion')
plt.show()
```

```python
datos2["prcp"].plot(linewidth=0.9,xlabel="Años",ylabel="mm",title="Evolución precipitación")
```
