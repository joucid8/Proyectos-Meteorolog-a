```python
datosMeteorologicos = pnd.read_csv("C:/Users/PC/Desktop/datosmeteorologicos.csv", index_col=1, parse_dates=True)  # utilizamos index_c
```

```python
datosMeteorologicos.index
```

```python
datosMeteorologicos['Año'] = datosMeteorologicos.index.year       #utilizamos la funcion .index (para ver la fecha( en mes,año y dia separado))                         
datosMeteorologicos['Mes'] = datosMeteorologicos.index.month
datosMeteorologicos['Dia'] = datosMeteorologicos.index.day 
datosMeteorologicos['Hora'] = datosMeteorologicos.index.hour
datosMeteorologicos['Minuto'] = datosMeteorologicos.index.minute
datosMeteorologicos['Segundo'] = datosMeteorologicos.index.second 
datosMeteorologicos.sample(45, random_state=0)  #para visualizar  45 datos
```
